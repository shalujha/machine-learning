# Generating Regression Data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(10,20,30) # this gives samples from uniform distribution which is not in ral life at all
print(x)
```

```python
plt.hist(x,)
```

```python
x=np.random.randn(3000) # this gives samples from Normal distribution which is not in ral life at all that is 0 mean and unit variance
print(x)
plt.hist(x,)
```

```python
u=50
std=1 #standard deviation simply means how much it has been deviated from mean
x=np.random.randn(300)*std+u# we have simply shifted the mean
noise=np.random.randn(300)
print(x)
plt.hist(x,)
```

```python
y=3*x+2+ noise
print(x.shape)
print(y.shape)
```

```python
plt.scatter(x,y,color='orange',label='3x+2')
plt.legend()
plt.show()
```

```python
# as  we can see from the graph that most of the points are at origin and as we move away from the mean point are not there
# after adding some noise they are not at the same position where they were supposed to be
```

#  Generating Classification data

```python
# Lets generate some examples of binary Classification, 
# Binary classification means it helps to identify whether the given data belongs to class 0 or class 1
# we will use multi variate normal distribution because we are talking about more than one feature , if we have single feature
# we will use normal distribution
```

```python
mean=np.array([2,4])
covariance=np.array([[1,0],[0,1]])
x1=np.random.multivariate_normal(mean,covariance,300)
plt.plot(x1)
```

```python
mean=np.array([2,4])
covariance=np.array([[1,0],[0,1]])
x1=np.random.multivariate_normal(mean,covariance,300)
plt.scatter(x1[:,0],x1[:,1])
plt.show()
```

```python
print(x1.shape)
```

```python
mean=np.array([2,4])
covariance=np.array([[1,0.8],[0.8,1]])
x1=np.random.multivariate_normal(mean,covariance,300)
plt.scatter(x1[:,0],x1[:,1])
plt.show()
```

```python
mean=np.array([2,4])
covariance=np.array([[1,-0.8],[-0.8,1]])
x1=np.random.multivariate_normal(mean,covariance,300)
plt.scatter(x1[:,0],x1[:,1])
plt.show()
```

```python
mean=np.array([2,4])
covariance=np.array([[1,0],[0,1]])
x1=np.random.multivariate_normal(mean,covariance,300)
mean2=np.array([-1,-3])
covariance2=np.array([[1,0],[0,2]])
x2=np.random.multivariate_normal(mean2,covariance2,400)
plt.scatter(x1[:,0],x1[:,1],label="Apple")
plt.scatter(x2[:,0],x2[:,1],marker='^',label="orange")
plt.xlabel("x1 (Size)")
plt.ylabel("x2 (sweetness)")
plt.title("Apples and Fruits graph")
plt.legend()
plt.show()
```

```python
print(x1.shape)
print(x2.shape)
```

```python
y=np.zeros(700)
y[:300]=1
print(y)
```

```python
# for combining 2 matrix , we will use stack function , which will take 2 arrays ans combine those 2 matrix either along axis 0 or axis 1
```

```python
X=np.vstack((x1,x2))
print(X.shape)
```

```python
plt.plot(X[:,0],X[:,1])
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
# since we cant distingguish between 2 features, we will be do this
for i in range(700):
    if y[i]==1:
        plt.scatter(X[i,0],X[i,1],color='green')
    else:
        plt.scatter(X[i,0],X[i,1],color='red')
plt.show()
        
```

```python
plt.scatter(X[:,0],X[:,1],c=y) # shortcut , for each different label, c parameter will assign a unique color
plt.show()
```
