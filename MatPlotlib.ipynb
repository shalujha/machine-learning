# Data Visualisation

## Matpotlib is a python library whic is used when we have to do some data visualisation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(10) # this arange function will give numbers from 0 to 9
print(x)
```

```python
plt.plot(x)
```

```python
# plot takes usually two parameters but when it takes only one parameter it means that one parameter is assigned to y and x is assigned the indices
# so the graph is between indices and values
```

```python
x=x**2
plt.plot(x)
```

```python
# so we can see that indices are same but values of y has differed which made changes in graph
```

```python
# to display the plot
plt.plot(x,color='green')
x2=x**3
plt.plot(x2,color='red')
plt.show()
```

```python
plt.scatter(x,x**2,color='orange',marker='^',label='stock Price') # label basically indicates what information the graph is giving to us
plt.legend() # # it will make a box for label
plt.xlabel("days") # it will basically give label to x axis
plt.ylabel("Year") # it will basically gives label to y axis
plt.title("Stock Price Prediction") # we can also give title to our graph
plt.show()
```

```python
plt.plot(x,label='2018',color='red')
plt.plot(x**2,label='2019',color='green')
plt.legend()
plt.show()
```

# BarGraph

```python
number=np.random.randint(0,10,5)
indices=np.arange(5)
number2=np.random.randint(0,10,5)
indices2=np.arange(5)
print(number)
print(indices)
```

```python
# numbers will become the height of bar and indices will become the width of bar
```

```python
plt.bar(indices,number,0.25,color='yellow',label='Gold') # 0.5 ha been used to decrease its width
plt.bar(indices2+0.25,number2,0.25,color='red',label='silver') # indices+0.5 is preventing overlapping
plt.legend()
plt.title("Gold Prediction")
plt.show()
```

# Pie Chart
- circular representation of data 

```python
labels=['coding','Algorithm','calculas','Algebra']
values=[90,80,70,87]
# it is not necessary for the sum to be 100,it will automatically calculated
plt.pie(values,labels=labels,radius=1)
plt.show()
```

```python
# to  highlight one particular part of pie chart , we will use explode function
plt.pie(values,labels=labels,radius=1,explode=(0,1,0,0))
plt.show()
```

```python
plt.pie(values,labels=labels,radius=2,explode=(0,1,0,0),shadow=True)
plt.show()
```
