# Algorithm:Linear Regression

```python
X=x
Y=y
def hypothesis(theta,X):
    return theta[0]+theta[1]*X
def error(theta,X,Y):
    m=X.shape[0]
    total_error=0
    for i in range(m):
        total_error+=(Y[i]-hypothesis(theta,X[i]))**2
    return total_error
def gradient(theta,Y,X):
    m=X.shape[0]
    grad=np.array([0.0,0.0])
    for i in range(m):
        grad[0]+=hypothesis(theta,X[i])-Y[i]
        grad[1]+=(hypothesis(theta,X[i])-Y[i])*X[i]
    return grad
def gradientdescent(X,Y,learning_rate,maxItr):
    theta=np.array([0.0,0.0])
    grad=np.array([0.0,0.0])
    for i in range(maxItr):
        grad=gradient(theta,X[i],Y[i])
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
    return theta
theta=gradientdescent(X,Y,learning_rate=0.001,maxItr=50)
print(theta)
        
        

    
```
