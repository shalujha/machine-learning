```python
from sklearn.datasets import make_regression,make_classification,make_blobs,make_moons
import matplotlib.pyplot as plt # it is used for plotting graphs and labelling
from sklearn.datasets import load_digits # powerful library used for data generation in machine learning
import numpy as np # library used for generating random numbers
```

```python
X,Y=make_regression(n_samples=400,n_features=1,n_informative=1,noise=0.8)
plt.plot(X,Y,label="Location")# basically what happens is make_regression returns two arguments X and Y means in X , there are samples of input values and in y , there are samples of output values
plt.xlabel("Price")
plt.ylabel("House")
plt.title("House Price")
plt.legend()
plt.show()
print(X.shape)
print(Y.shape)
```

```python
XC,YC=make_classification(n_samples=400,n_features=2,n_informative=2,n_redundant=0)
plt.scatter(XC[:,0],XC[:,1],c=YC) # XC  is the generated samples and YC is the integer label of class membership of each sample
plt.title("Classification")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
```

```python
XB,YB=make_blobs(n_samples=400,n_features=3,centers=5,random_state=5)
plt.plot(XB,YB)
# XB is the generated samples and YB is the cluster membership of each sample
```

```python
plt.scatter(XB[:,0],XB[:,1],c=YB)
```

```python
XM,YM=make_moons(n_samples=100,noise=None,random_state=5) # XM is the generated samples and YM is the class membership of each samples
```

```python
plt.scatter(XM[:,0],XM[:,1],c=YM)
```
